```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
```

```python
training = pd.read_csv("train.csv")
```

```python
training
```

```python
training.drop(["Name","Ticket","PassengerId","Cabin"],axis=1,inplace=True)
```

```python
age = training["Age"].mean()
training["Age"].fillna(29.0,inplace = True)
training.fillna(0)
```

```python
sb.countplot(training["Survived"])
```

```python
sb.countplot(training[training["Sex"] == "female"]["Survived"],label="survived female")
```

```python
train_x,train_y = training[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare','Embarked']].values,training["Survived"].values.reshape(-1,1)
```

```python
train_x
```

```python
train_y
```

```python
for i in range(len(train_x)):
    if train_x[i,1] == "male":
        train_x[i,1] = 0
    if train_x[i,6] == "S":
        train_x[i,6] = 0
    if train_x[i,6] == "C":
        train_x[i,6] = 1
    if train_x[i,6] == "Q":
        train_x[i,6] = 2
    if train_x[i,1] == "female":
        train_x[i,1] = 1

        
train_x[61,6] = 0
train_x[829,6] = 0
```

```python
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
train_x = scale.fit_transform(train_x)
```

```python
np.where(np.isnan(train_x))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=200)
```

```python
model.fit(train_x,train_y)
```
