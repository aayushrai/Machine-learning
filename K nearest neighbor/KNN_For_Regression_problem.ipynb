```python
x = data[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']].values
x
```

```python
y = data['Price'].values
y
```

```python
def eulidean_Distance(old_x,new_x,y,k):
    distance = list([np.sqrt(sum((new_x[col]-old_x[row,col])**2 for col in range(old_x.shape[1]))),y[row]]for row in range(old_x.shape[0]))
    distance = sorted(distance,key= lambda x:x[0])
    return np.array(distance[:k])
    
```

```python
new =[7.95454586e+04, 5.68286132e+00, 7.00918814e+00, 4.09000000e+00,2.30868005e+04]
min_k_distance = eulidean_Distance(x,new,y,3)
min_k_distance
```

```python
prediction = np.mean(min_k_distance[:,1])      # we take mean of top k nearest value in our case k = 3
prediction
```
